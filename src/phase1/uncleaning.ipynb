```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("Kaggle_Dataset_Perth_Housing.csv")
dataset
```

```python
dataset.shape
```

```python
dataset["DATE_SOLD"] = dataset["DATE_SOLD"].str.replace("\r","")
```

```python
#Duplicating rows and randomly adding in the dataset

print("\nShape : ", dataset.shape)
rows_to_duplicate = np.random.randint(0,dataset.shape[0],size=1000)
selected_rows = dataset.iloc[rows_to_duplicate]
dataset = pd.concat([dataset,selected_rows],ignore_index=True)
dataset = dataset.sample(frac=1).reset_index(drop=True)
print("\nShape after duplicating rows : ", dataset.shape,"\n")
print(dataset)
```

```python
print("NA values : \n")
print(dataset.isna().sum())

#Creating Empty Values

create_na_columns = ['PRICE', 'BEDROOMS', 'BATHROOMS', 'GARAGE', 'LAND_AREA', 'FLOOR_AREA', 'BUILD_YEAR', 'CBD_DIST', 'NEAREST_STN_DIST', 'NEAREST_SCH_DIST']
for create_na_col in create_na_columns:
    rand = np.random.randint(2000,3001)
    
    indices = np.random.choice(dataset.shape[0], size=rand, replace=False)
    dataset.loc[indices,create_na_col] = np.nan

print("\nNA values after transformation :\n")
print(dataset.isna().sum())
dataset
```

```python
dataset.info()
dataset.to_csv("uncleaned_dataset.csv", index=False)
```
